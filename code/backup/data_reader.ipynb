```python
import numpy as np
import os# to handle files in the folder
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filepath = "C:/Users/Administrator/Desktop/cavity_magnonics_TP4/data/"

def viewer(filename, x_axis = "field", sweep=1): 
    PcFile=filepath+filename
    isFile = os.path.isfile(PcFile)
    if isFile:
        data = pd.read_csv(PcFile,sep=';')
        time = tuple(data.Time)
        disp = tuple(data.Display)
        avg = float(data.Averages[0])
        freq = tuple(map(float, str(data["Frequencies[Hz]"][0]).split(",")))[::-1]
        field = tuple(map(float, data["Field[mT]"]))
        trace = tuple(data.RealandImaginary)
        re = []
        im = []
        for i in range(len(disp)):
            row = []
            row = tuple(map(float, trace[i].split(",")))[::-1]
            re.append(row[::2])
            im.append(row[1::2])
        re = np.array(re)
        im = np.array(im)
        amp = (re**2 + im**2)**0.5
        
        if x_axis == "time":
            x = time    
        else:
            x = field
        
        step_x = int(len(x) / 6) # step between consecutive labels
        x_positions = np.arange(0.1,len(x),step_x) # pixel count at label position
        x_labels = x[4::step_x] # labels you want to see 
        x_labels = ['%.3f'%(i) for i in x_labels]
        step_y = int(len(freq) / 6)
        y_positions = np.arange(0, len(freq), step_y)
        y_labels = freq[::step_y]
        y_labels = ['%.3f'%(i/1e9) for i in y_labels]
        plt.close()
        plt.figure(figsize=(15,15))
        plt.imshow(amp[::sweep].T, cmap="plasma", aspect = len(x)/len(freq)) #color is "plasma"
        plt.colorbar()
        plt.xticks(x_positions[::sweep], x_labels, fontsize=15)
        plt.xlabel("field[mT]", fontsize=20)
        plt.ylabel("frequency[GHz]", fontsize=20)
        plt.yticks(y_positions, y_labels, fontsize=15)
        plt.show()
    else:
        print("file "+filename+" does not exist :/")
        return 0
```

```python
viewer("test_yig_series-30dB.csv")
```
